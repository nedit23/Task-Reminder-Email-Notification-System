```python
import tkinter as tk
from tkinter import messagebox
import json
from datetime import datetime, timedelta
import smtplib
from email.mime.text import MIMEText

FILE_NAME = "task1.json"

EMAIL_ADDRESS = "user@gmail.com"
EMAIL_PASSWORD = "user_app_password"

# ---------------- FILE ---------------- #
def load_tasks():
    try:
        with open(FILE_NAME, "r") as file:
            return json.load(file)
    except:
        return []

def save_tasks(tasks):
    with open(FILE_NAME, "w") as file:
        json.dump(tasks, file, indent=4)

# ---------------- EMAIL ---------------- #
def send_email(to_email, task_name, deadline):
    try:
        subject = "Task Reminder"
        
        body = f"""Your task "{task_name}" is due soon.
Please ensure that the task is completed before the deadline.

Deadline: {deadline}
"""

        msg = MIMEText(body)
        msg["Subject"] = subject
        msg["From"] = EMAIL_ADDRESS
        msg["To"] = to_email

        server = smtplib.SMTP("smtp.gmail.com", 587)
        server.starttls()
        server.login(EMAIL_ADDRESS, EMAIL_PASSWORD)
        server.send_message(msg)
        server.quit()

        print("Email sent successfully!")

    except Exception as e:
        print("Error:", e)

# ---------------- FUNCTIONS ---------------- #
def add_task():
    name = entry_name.get()
    desc = entry_desc.get()
    deadline = entry_deadline.get()
    email = entry_email.get()

    if not name or not deadline or not email:
        messagebox.showerror("Error", "Fill all required fields")
        return

    task = {
        "name": name,
        "description": desc,
        "deadline": deadline,
        "email": email
    }

    tasks = load_tasks()
    tasks.append(task)
    save_tasks(tasks)

    messagebox.showinfo("Success", "Task Added!")

    entry_name.delete(0, tk.END)
    entry_desc.delete(0, tk.END)
    entry_deadline.delete(0, tk.END)
    entry_email.delete(0, tk.END)

    view_tasks()

def view_tasks():
    listbox.delete(0, tk.END)
    tasks = load_tasks()
    for i, task in enumerate(tasks):
        listbox.insert(tk.END, f"{i+1}. {task['name']} | {task['deadline']}")

def check_deadlines():
    tasks = load_tasks()
    now = datetime.now()

    for task in tasks:
        deadline = datetime.strptime(task["deadline"], "%Y-%m-%d")
        diff = deadline - now

        if timedelta(hours=0) < diff <= timedelta(hours=48):
            send_email(task["email"], task["name"], task["deadline"])

    messagebox.showinfo("Done", "Checked all deadlines!")

def delete_task():
    selected = listbox.curselection()
    if not selected:
        return

    index = selected[0]
    tasks = load_tasks()
    tasks.pop(index)
    save_tasks(tasks)

    view_tasks()
    messagebox.showinfo("Deleted", "Task Removed!")

# ---------------- GUI ---------------- #
root = tk.Tk()
root.title("✨ Task Reminder System")
root.geometry("520x550")
root.configure(bg="#1e1e2f")   # Dark background

# Title
title = tk.Label(root, text="Task Reminder", font=("Arial", 18, "bold"),
                 bg="#1e1e2f", fg="#00d4ff")
title.pack(pady=10)

# Frame for inputs
frame = tk.Frame(root, bg="#2b2b3c", padx=10, pady=10)
frame.pack(pady=10)

def create_label(text):
    return tk.Label(frame, text=text, bg="#2b2b3c", fg="white", font=("Arial", 10))

def create_entry():
    return tk.Entry(frame, width=35, bg="#3c3f58", fg="white", insertbackground="white")

create_label("Task Name").grid(row=0, column=0, sticky="w")
entry_name = create_entry()
entry_name.grid(row=0, column=1, pady=5)

create_label("Description").grid(row=1, column=0, sticky="w")
entry_desc = create_entry()
entry_desc.grid(row=1, column=1, pady=5)

create_label("Deadline (YYYY-MM-DD)").grid(row=2, column=0, sticky="w")
entry_deadline = create_entry()
entry_deadline.grid(row=2, column=1, pady=5)

create_label("Email").grid(row=3, column=0, sticky="w")
entry_email = create_entry()
entry_email.grid(row=3, column=1, pady=5)

# Button style
def styled_button(text, command, color):
    return tk.Button(root, text=text, command=command,
                     bg=color, fg="white",
                     font=("Arial", 10, "bold"),
                     width=20, relief="flat")

styled_button("➕ Add Task", add_task, "#4CAF50").pack(pady=5)
styled_button("📋 View Tasks", view_tasks, "#2196F3").pack(pady=5)
styled_button("⏰ Check Deadlines", check_deadlines, "#ff9800").pack(pady=5)
styled_button("❌ Delete Task", delete_task, "#f44336").pack(pady=5)

# Task List
listbox = tk.Listbox(root, width=60, height=10,
                     bg="#2b2b3c", fg="white",
                     selectbackground="#00d4ff",
                     font=("Arial", 10))
listbox.pack(pady=15)

root.mainloop()
```
